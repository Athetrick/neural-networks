Загрузка набора данных о хронической болезни почек (Chronic Kidney Disease) из репозитория UCI

```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
# fetch dataset
chronic_kidney_disease = fetch_ucirepo(id=336)

# data (as pandas dataframes)
X = chronic_kidney_disease.data.features
y = chronic_kidney_disease.data.targets
df = pd.concat([X, y], axis=1)
# metadata
# print(chronic_kidney_disease.metadata)

# variable information
# print(chronic_kidney_disease.variables)
```

# Предварительная оработка данных

Анализ и вывод информации о пропущенных значениях (NaN) в загруженном наборе данных о хронической болезни почек. Определим количество пропущенных значений в каждом признаке.

```python
print("Пропущенные значения в датасете:")
print(df.isnull().sum())
```

Разделяем переменные на числовые и категориальные. Визуализируем распределение числовых показателей в наборе.

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Определяем числовые и категориальные признаки
num_cols = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']
cat_cols = list(set(df.columns) - set(num_cols) - {'class'})

# Анализ распределения числовых признаков перед заполнением
for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Распределение {col}')
    # plt.show()
```

**1. Обработка пропущенных данных**

Для признаков, в которых было найдено 1-2 пропущенных значения, было принято решение удалить эти строки, поскольку их пропуски несущественны и их заполнение не имеет смысла. Признаки, в которых пропущены значения, но их было больше, были обработаны с использованием методов заполнения:
Для **категориальных признаков** (например, htn, dm и т. д.) пропуски были заполнены модой, то есть наиболее часто встречающимся значением.
Для **числовых признаков** использовался метод заполнения значений на основе ближайших соседей (KNN), где пропущенные данные заполнялись с помощью 5 ближайших значений.


```python
# Удаление строк, для признаков у которых 1-2 пропущенное значение
cols_to_drop_na = ['htn', 'dm', 'cad', 'appet', 'pe', 'ane']
df = df.dropna(subset=cols_to_drop_na)
from sklearn.impute import SimpleImputer, KNNImputer
# Для категориальных данных - наиболее частое значение
cat_imputer = SimpleImputer(strategy='most_frequent')
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])
# Для номинальных переменных
knn_imputer = KNNImputer(n_neighbors=5)
df[num_cols] = knn_imputer.fit_transform(df[num_cols])

print("Пропущенные значения в датасете:")
print(df.isnull().sum())
```

В результате был получен Dataset без пропусков

**2. Кодирование признаков в числовые значения для последующего анализа и моделирования**

После обработки пропусков проведем кодирование признаков в числовой формат.

```python

binary_cols = ['htn', 'dm', 'cad', 'pe', 'ane']
for col in binary_cols:
    df[col] = df[col].map({'yes': 1, 'no': 0, '\tno':0})

cols_2 = ['rbc', 'pc']
for col in cols_2:
    df[col] = df[col].map({'abnormal': 1, 'normal': 0})

cols_3 = ['ba', 'pcc']
for col in cols_3:
    df[col] = df[col].map({'present': 1, 'notpresent': 0})

df['appet'] = df['appet'].map({'good': 1, 'poor': 0})
df['class'] = df['class'].map({'ckd': 1, 'notckd': 0,'ckd\t': 0})
```

Для бинарных признаков (таких как htn, dm, cad, и т. д.) использовалась бинарная кодировка: значения yes и no были заменены на 1 и 0, соответственно. Для других категориальных признаков, таких как rbc и pc, значения abnormal и normal были заменены на 1 и 0. Признак class был преобразован в бинарный формат с метками ckd (1) и notckd (0).

```python
print("Пропущенные значения в датасете:")
print(df.isnull().sum())
```

**3. Анализ корреляционной матрицы числовых признаков**

Для дальнейшего анализа взаимосвязей между числовыми построим корреляционную матрицу.

```python
# Анализ корреляции
plt.figure(figsize=(12, 8))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

В данных нет сильно связанных признаков, каждый признак в определенной степени независим от других.

**4. Масштабирование числовых признаков**

Для приведения числовых признаков к общему масштабу используем масштабирование с использованием стандартизации.

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
# Масштабирование числовых признаков
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

Метод StandardScaler привел данные к распределению с нулевым средним и единичной дисперсией, что важно для последующего моделирования.

**5. Анализ и отбор признаков по дисперсии**


Для уменьшения размерности набора данных выберем признаки с дисперсией больше или равной 0.05.

```python
variances = df.drop(columns=['class']).var()
variance_report = pd.DataFrame({
    'Признак': variances.index,
    'Дисперсия': variances.values,
    'Будет отобран': variances.values >= 0.05
})
print(variance_report)
selected_features = variances[variances >= 0.05].index
# Оставляем только отобранные признаки и целевую переменную 'class'
df = df[list(selected_features) + ['class']]
print(df)
```

Так как признаки с низкой дисперсией (например, sg) не вносят значимого вклада в анализ и могут быть исключены из дальнейшего рассмотрения.
